##### 1.❓Are there better or worse days/times in terms of sales? Plot a chart to show the result

```python
# Extracting dates from transaction date
coffee['transaction_date']
```

```python
# Creating time categories function to return Morning, Noon, Evenging, Night based on transaction_time
hours = pd.to_datetime(coffee['transaction_time'], format='%H:%M:%S').dt.hour

def time_category(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Noon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'



# use apply to apply the time category into a new column called time_category
coffee['time_category'] = hours.apply(time_category)
coffee.head()
```

```python
# Are there better or worse days/times in terms of sales?
days_times = coffee.groupby(["transaction_day", "time_category"])["line_item_amount"].sum().sort_values(ascending = False)
days_times.plot(kind = 'bar', title = 'Days and times in terms of sales');

# Write your findings

# mornings have a higher number of sales than other times
```
